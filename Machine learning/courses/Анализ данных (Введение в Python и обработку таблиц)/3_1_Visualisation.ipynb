# Визуализация

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('students.csv')
```

## Столбчатые диаграммы (распределения)

```python
sns.displot(df, x='Growth')
```

```python
sns.displot(df, x='Growth', kind='kde')
```

## Облака точек

```python
sns.scatterplot(data=df, x='Growth', y='Weight', hue='Sex')
```

## Категориальные признаки

```python
sns.countplot(data=df, x='Sex')
```

```python
sns.countplot(data=df, x='Sex', hue='Animal')
```

## Попарная зависимость признаков

```python
df_cut = df[['Age', 'Growth', 'Weight', 'Shoe size', 'Sex']]
```

```python
sns.pairplot(df_cut, hue='Sex')
```
