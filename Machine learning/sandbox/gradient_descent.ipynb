## Реализация градиентного спуска

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Генерация случайного набора данных (линейная зависимость ax + b + шум)

# Коэффициенты и диапазон x подобраны так, чтобы получился красивый изгиб
a3 = 1.1
a2 = -4
a1 = -10
a0 = 5

x = np.arange(-4, 6, 0.1)
y = a3 * x ** 3 + a2 * x ** 2 + a1 * x + a0 + np.random.uniform(-10,10,100)
```

```python
plt.plot(y)
```

```python
# b1 и b0 - это текущие коэффициенты нашей линейной регрессии.
# Мы должны их максимально приблизить к a1 и a0
# Но сначала они случайные
b3 = np.random.random()
b2 = np.random.random()
b1 = np.random.random()
b0 = np.random.random()
print (f'b3 = {b3:.3f}, b2 = {b2:.3f}, b1 = {b1:.3f}, b0 = {b0:.3f}')
```

```python
# Шаг обучения:
lr = 0.000001
```

```python
# Основной цикл обучения:
for i in range(10000):  # 10_000 шагов

    # Прогноз модели
    y_hat = b3 * x ** 3 + b2 * x ** 2 + b1 * x + b0

    # Функция потерь  - MSE (mean squared error)
    loss = ((y - y_hat) ** 2).sum()
    print(f'MSE_loss: {loss:.2f}')

    # При использовании функции потерь MSE получим градиент
    dl_db0 = (2 * (y_hat - y)).sum()      # Частная производная лосс-функции по коэффициенту b0
    dl_db1 = (2 * (y_hat - y) * x).sum()  # Частная производная лосс-функции по коэффициенту b1
    dl_db2 = (2 * (y_hat - y) * x ** 2).sum()  # Частная производная лосс-функции по коэффициенту b2
    dl_db3 = (2 * (y_hat - y) * x ** 3).sum()  # Частная производная лосс-функции по коэффициенту b3
    grad = [dl_db0, dl_db1, dl_db2, dl_db3]  # Градиент

    # Изменяю коэффициенты в направлении антиградиента с шагом обучения lr
    b0 -= lr * dl_db0
    b1 -= lr * dl_db1
    b2 -= lr * dl_db2
    b3 -= lr * dl_db3
```

```python
# Финальное предсказание модели
y_hat = b3 * x ** 3 + b2 * x ** 2 + b1 * x + b0
```

```python
plt.plot(y)
plt.plot(y_hat)
```
