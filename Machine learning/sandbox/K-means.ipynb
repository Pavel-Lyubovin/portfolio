# Модели кластеризации методом K-средних

```python
from sklearn.datasets import make_blobs
```

```python
raw_data = make_blobs(
    n_samples = 200, 
    n_features = 2, 
    centers = 4, 
    cluster_std = 1.8
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Визуализация датасета

```python
plt.scatter(raw_data[0][:,0], raw_data[0][:,1])
```

```python
# Раскрасим кластеры в разные цвета
plt.scatter(raw_data[0][:,0], raw_data[0][:,1], c=raw_data[1]);
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4)
```

```python
model.fit(raw_data[0])
```

```python
# Это предсказания нашей модели (к какому классу относится каждая точка)
model.labels_
```

```python
# Это предсказания нашей модели (центры кластеров)
model.cluster_centers_
```

## Визуализация точности предсказаний модели

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('Наши предсказания')
ax1.scatter(raw_data[0][:,0], raw_data[0][:,1],c=model.labels_)
ax2.set_title('Реальные значения')
ax2.scatter(raw_data[0][:,0], raw_data[0][:,1],c=raw_data[1]);
```
